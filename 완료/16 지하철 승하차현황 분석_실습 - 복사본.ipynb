## 승차/하차 테이블 분리

### 승차 테이블 만들기

```python
# 공통 컬럼
df1 = df.iloc[:,0:3]

# 승차 컬럼만 가져오기
df2 = df.iloc[:,3::2]
df2.columns = df2.columns.str.split(' ').str[0]
```

```python
df2
```

```python
# 공통컬럼과 승차컬럼 연결하기
df_in = pd.concat([df1,df2],axis=1)
df_in
```

## 하차 테이블 만들기

```python
# 공통 컬럼
df3 = df.iloc[:,:3]

# 하차 컬럼만 가져오기
df4 = df.iloc[:,4::2]
df4.columns = df4.columns.str.split(' ').str[0]
df4
```

```python
# 공통컬럼과 하차컬럼 연결하기
df_out = pd.concat([df3,df4],axis=1)
```
