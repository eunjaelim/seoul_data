# 삼겹살 가격 분석

## 삼겹살 데이터
* 특정 문자열이 포함되어있는지 확인 : .str.contains(문자열)

```python
df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06')]
```

```python
# 2021-06 데이터 이용
df_sam = df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06') & (df['실판매규격'].str.contains('600g'))]
```

```python
# 삼겹살 600g의 평균 가격은?
df_sam['가격(원)'].mean()
```

```python
# 삼겹살 600g의 최고 가격은?
```

```python
df_sam['가격(원)'].max()
```

```python
# 삼겹살 600g의 최저 가격은?
df_sam['가격(원)'].min()
```

```python
df_sam.dtypes
```

```python
df_sam[df_sam['가격(원)'] < 5000]
```

## 우리동네 삼겹살 가격

```python
gu = input('구이름:')
```

```python
# 우리구 삼겹살 가격
df_sam_gu = df_sam[df_sam['자치구 이름']==gu][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
```

```python
# 시각화
plt.rc('font', family='Malgun Gothic')
x = df_sam_gu['시장/마트 이름']
y = df_sam_gu['가격(원)']
plt.scatter(x,y)
plt.title(gu+ ' 삼겹살 가격')
plt.show()
```

## 마트 지점별 삼겹살 가격

```python
mart = input('시장/마트이름:')
```

```python
# 마트 지점별 삼겹살 가격
df_sam_mart = df_sam[df_sam['시장/마트 이름'].str.contains(mart)][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
df_sam_mart
```

```python
# 시각화
x = df_sam_mart['시장/마트 이름']
y = df_sam_mart['가격(원)']
plt.scatter(x,y)
plt.rc('font', family='Malgun Gothic')
plt.grid(True)
plt.title(mart+' 삼겹살 가격')
plt.xticks(rotation=45)
plt.show()
```

# 달걀 가격 분석

## 달걀 데이터

```python
df
```

```python
# 2021-06 데이터 이용
df_egg = df[(df['품목 이름'].str.contains('달걀')) & (df['년도-월'] == '2021-06')][['시장/마트 이름','품목 이름','실판매규격','가격(원)','자치구 이름']].drop_duplicates()
df_egg
```

```python
# 달걀 30구 평균 가격
df_egg = df_egg[(df_egg['품목 이름'].str.contains('30개')) & (df_egg['가격(원)'] > 0)]
df_egg['가격(원)'].mean()
```

```python
# 달걀 최고 가격
```

```python
df_egg[df_egg['가격(원)'] > 20000]
```

```python
df_egg[df_egg['가격(원)'] < 6000]
```

## 우리동네 달걀 가격

```python
gu = input('구이름:')
```

```python
# 우리구 달걀 가격
df_egg_gu = df_egg[df_egg['자치구 이름'].str.contains(gu)][['시장/마트 이름','가격(원)']].drop_duplicates()
df_egg_gu
```

```python
df_egg_gu = df_egg_gu.sort_values('가격(원)')
```

```python
# 시각화
x = df_egg_gu['시장/마트 이름']
y = df_egg_gu['가격(원)']
plt.scatter(x,y)
plt.grid(True)
plt.title(gu+' 달걀가격(30구)')
plt.show()
```
