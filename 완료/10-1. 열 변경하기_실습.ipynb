```python
import pandas as pd
df = pd.read_csv('data/scores.csv')
df.head()
```

# 열 추가/수정하기
* **데이터프레임[컬럼] = 추가할데이터**  
* **데이터프레임[컬럼] = 수정할데이터**  
컬럼이 존재하면 추가, 존재하지 않으면 수정된다.

## 열 추가하기

```python
# 학생 번호 추가하기 (1부터 시작하여 1씩 증가)
```

```python
len(df)
```

```python
df['no'] = range(1,31)
```

```python
df
```

```python
# sum 추가
```

```python
df['sum']= df['kor']+df['eng']+df['math']
```

```python
df
```

## 열 수정하기

```python
# 학생 번호 수정하기 (100부터 시작하여 1씩 증가)
```

```python
df['no']= df['no']+99
```

```python
df
```

# 열 삭제하기
* 데이터프레임.**drop**(columns=삭제할컬럼리스트, inplace=True)
* 존재하지 않는 열은 삭제할 수 없다.

```python
# no, sum 컬럼 삭제하기
```

```python
df.drop(columns=['no','sum'],inplace=True)
```

```python
df
```

# 컬럼명 바꾸기

## 컬럼명 한번에 바꾸기
* 데이터프레임.**columns**  = 컬럼명리스트           
  컬럼명리스트의 항목 수는 컬럼 수와 동일해야한다.

```python
#'이름','국어','영어','수학'
df.columns=['이름','국어','영어','수학']
```

```python
df
```

```python
# 전체컬럼수와 컬럼명 리스트의 항목 수가 다르면 error
```
